```python
import matplotlib.pyplot as plt
import scienceplots
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#Inputting traffic result's excel
#It includes 4 data:
# 1. primary survey traffic data
# 2. TraffiX's result using Indonesian BPR function 
# 3. TraffiX's result using US BPR Function.
# 4. PTV VISUM's result
var2 = pd.read_excel("Plot regresi visum.xlsx")
display(var2)
```

```python
#We plot the results using MatPlotLib
plt.style.use(['science','ieee','no-latex'])
#values
x_values2 = list(var2['Volume Survei'])
Indo = list(var2['Volume BPR Indo'])
USA = list(var2['Volume BPR USA'])
Visum = list(var2['Volume Visum'])


#plotting
fig, axs = plt.subplots(nrows=1,ncols=3,
                        constrained_layout=True, figsize=(6, 1.8))
axs[0].scatter(x_values2, Indo, s=2)
axs[0].axline(xy1=(0, 0), slope=1.4209, label=f'$y = {1.354}x$',linestyle='dashed')
axs[0].set_xlabel('Observed traffic flow (pcu/hour)')
axs[0].set_ylabel('Modeled traffic flow (pcu/hour)')
axs[0].set_title('(a)')
axs[0].set_ylim((0,12000))
axs[1].scatter(x_values2, USA, s=2)
axs[1].axline(xy1=(0, 0), slope=1.4606, label=f'$y = {1.6266}x$',linestyle='dashed')
axs[1].set_xlabel('Observed traffic flow (pcu/hour)')
axs[1].set_ylabel('Modeled traffic flow (pcu/hour)')
axs[1].set_title('(b)')
axs[1].set_ylim((0,12000))
axs[2].scatter(x_values2, Visum, s=2)
axs[2].axline(xy1=(0, 0), slope=1.1116, label=f'$y = {1.4353}x$',linestyle='dashed')
axs[2].set_xlabel('Observed traffic flow (pcu/hour)')
axs[2].set_ylabel('Modeled traffic flow (pcu/hour)')
axs[2].set_title('(c)')
axs[2].set_ylim((0,12000))
```
